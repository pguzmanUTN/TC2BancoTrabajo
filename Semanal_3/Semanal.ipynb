# Insciso 3 de Simulacion de Transferencia normalizada

En este codigo de paso cree un script donde me calcula directamente la transferencia normalizada apartir de los datos pedidos en la plantilla

```python


##Coeficientes obtenidos algebraicamente
a=1.2526
b=1.2526
c=1.569
#En este caso dio la casualidad de que Q=1 para la parte de orden 2, por eso quedaron a y b iguales, ademas a=w0 y c=w0 al cuadrado
#Este omcosp quedo muy simetrico


def mostrar_transferencia_latex(num_coefs, den_coefs, nombre='T'):
    num = Poly(num_coefs, s).as_expr()
    den = Poly(den_coefs, s).as_expr()

    fraccion = simplify(num / den)

    eq = symbols(nombre + '(s)')
    latex_expr = f"{nombre}(s) = \\frac{{{latex(num)}}}{{{latex(den)}}}"

    display(Math(latex_expr))

def butterworth_max_planicidad(alfa_max, alfa_min, wp, ws):
    ep2 = 10**(alfa_max / 10) - 1
    A2 = 10**(alfa_min / 10) - 1
    ep=np.sqrt(ep2)
    n = np.ceil(np.log10(A2 / ep2) / (2 * np.log10(ws / wp))).astype(int)

    b, a = signal.butter(N=n, Wn=wp*(ep**(-1/n)), btype='low', analog=True)

    # Print transferencia
    H = signal.TransferFunction(b, a)
    print(f"Orden del filtro: {n}")
    print("Numerador:", H.num)
    print("Denominador:", H.den)

    return H

H = butterworth_max_planicidad(1, 12, 1, 2)


# Definición de la función de transferencia
numerador1 = [a] 
denominador1 = [1,a]  
numerador2 = [c] 
denominador2 = [1,b,c]  
my_tf = TransferFunction( np.polymul(numerador1,numerador2), np.polymul(denominador1,denominador2))




# Define el rango de frecuencias para la simulación
frecuencias = np.logspace(-2, 2, 500)   

# Calcular la respuesta en frecuencia
w, mag, fase = signal.bode(my_tf, frecuencias)
w2, mag2, fase2 = signal.bode(H, frecuencias)

# Graficar la magnitud
plt.figure(figsize=(10, 6))

# Magnitud (en dB)
plt.subplot(2, 1, 1)
plt.semilogx(w, 10**(mag/20),label='my_tf')  
plt.semilogx(w2, 10**(mag2/20), color='red',linestyle='--',label='H(s)')
plt.title('Respuesta en Magnitud y Fase de la Transferencia')
plt.legend(loc='best')
plt.ylabel('Modulo (Amplitud)')
plt.grid()

# Fase
plt.subplot(2, 1, 2)
plt.semilogx(w, fase,label='my_tf')
plt.semilogx(w2, fase2, color='red',linestyle='--',label='H(s)') 
plt.legend(loc='best')
plt.xlabel('Frecuencia (rad/s)')
plt.ylabel('Fase (grados)')
plt.grid()

plt.show()

plt.close('all')

pzmap(my_tf,annotations=False ,fig_id=2) #S plane pole/zero plot
pzmap(H,annotations=False ,fig_id=1) #S plane pole/zero plot
```

```python
mostrar_transferencia_latex(H.num, H.den, nombre='H')
```

```python
mostrar_transferencia_latex(my_tf.num, my_tf.den, nombre='mytf')
```
