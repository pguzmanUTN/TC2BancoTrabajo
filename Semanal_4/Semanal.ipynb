```python
z,p,k = sig.besselap(1, norm='delay')
num, den = sig.zpk2tf(z,p,k)
this_sos = tf2sos_analog(num, den)
print( 'Transferencia de Bessel n=1')
print( '------------------------')
pretty_print_lti( num, den )
pretty_print_SOS( this_sos, mode='omegayq' )

z,p,k = sig.besselap(2, norm='delay')
num, den = sig.zpk2tf(z,p,k)
this_sos = tf2sos_analog(num, den)
print( 'Transferencia de Bessel n=2')
print( '------------------------')
pretty_print_lti( num, den )
pretty_print_SOS( this_sos, mode='omegayq' )

z,p,k = sig.besselap(3, norm='delay')
num, den = sig.zpk2tf(z,p,k)
this_sos = tf2sos_analog(num, den)
print( 'Transferencia de Bessel n=3')
print( '------------------------')
pretty_print_lti( num, den )
pretty_print_SOS( this_sos, mode='omegayq' )

z,p,k = sig.besselap(4, norm='delay')
num, den = sig.zpk2tf(z,p,k)
this_sos = tf2sos_analog(num, den)
print( 'Transferencia de Bessel n=4')
print( '------------------------')
pretty_print_lti( num, den )
pretty_print_SOS( this_sos, mode='omegayq' )





```
